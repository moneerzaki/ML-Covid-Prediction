```python
### Don't mind about this
import warnings
warnings.filterwarnings('ignore')
###

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import math
```

```python
df = pd.read_csv('Covid.csv')
```

```python
df.head()
```

```python
training_data = df.drop('CLASIFFICATION_FINAL', axis=1)
training_label = df['CLASIFFICATION_FINAL']
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, roc_curve, precision_recall_curve, make_scorer, confusion_matrix
scoring_metrics = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score),
    'roc_auc_score': make_scorer(roc_auc_score),
}
    # 'roc_curve': make_scorer(roc_curve),
    # 'precision_recall_curve': make_scorer(precision_recall_curve)
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
folds = 10
LR_model = LogisticRegression(random_state=0, max_iter=1500, n_jobs=8)
LR_scores = cross_validate(LR_model, training_data, training_label, cv=folds, scoring=scoring_metrics)
```

```python
for i in range(folds):
    print(f"Accuracy for fold {i} = {LR_scores['test_accuracy'][i]}")
    print(f"Precision for fold {i} = {LR_scores['test_precision'][i]}")
    print(f"Recall for fold {i} = {LR_scores['test_recall'][i]}")
    print(f"F1-score for fold {i} = {LR_scores['test_f1'][i]}")    
    print(f"AUC for fold {i} = {LR_scores['test_roc_auc_score'][i]}")
    print("\n")

print("\nFinal Metrics>>")
print(f"Accuracy = {sum(LR_scores['test_accuracy'])/folds}")
print(f"Precision = {sum(LR_scores['test_precision'])/folds}")
print(f"Recall = {sum(LR_scores['test_recall'])/folds}")
print(f"F1-score = {sum(LR_scores['test_f1'])/folds}")
print(f"AUC = {sum(LR_scores['test_roc_auc_score'])/folds}")
```

# Neural Network

```python
from sklearn.neural_network import MLPClassifier
folds = 10
NN_model = MLPClassifier(random_state=1, max_iter=300, hidden_layer_sizes=[10,10], learning_rate='adaptive')
NN_scores = cross_validate(NN_model, training_data, training_label, cv=folds, scoring=scoring_metrics)
```

```python
for i in range(folds):
    print(f"Accuracy for fold {i} = {NN_scores['test_accuracy'][i]}")
    print(f"Precision for fold {i} = {NN_scores['test_precision'][i]}")
    print(f"Recall for fold {i} = {NN_scores['test_recall'][i]}")
    print(f"F1-score for fold {i} = {NN_scores['test_f1'][i]}")    
    print(f"AUC for fold {i} = {NN_scores['test_roc_auc_score'][i]}")
    print("\n")

print("\nFinal Metrics>>")
print(f"Accuracy = {sum(NN_scores['test_accuracy'])/folds}")
print(f"Precision = {sum(NN_scores['test_precision'])/folds}")
print(f"Recall = {sum(NN_scores['test_recall'])/folds}")
print(f"F1-score = {sum(NN_scores['test_f1'])/folds}")
print(f"AUC = {sum(NN_scores['test_roc_auc_score'])/folds}")
```
