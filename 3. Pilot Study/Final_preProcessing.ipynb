```python
columns_to_replace = df.columns.difference(['MEDICAL_UNIT', 'AGE', 'CLASIFFICATION_FINAL'])
df[columns_to_replace] = df[columns_to_replace].replace(2,0)
df.head()
```

```python
units = pd.get_dummies(df['MEDICAL_UNIT'], prefix='MEDICAL_UNIT', dtype = int)
df = df.drop('MEDICAL_UNIT', axis= 1)
df = pd.concat([df, units], axis= 1)
print(df.info())
```
