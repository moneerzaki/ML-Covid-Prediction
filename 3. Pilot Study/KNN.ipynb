```python
### Don't mind about this
import warnings
warnings.filterwarnings('ignore')
###

%matplotlib inline
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import matplotlib.pyplot as plt
import math
import cuml
```

```python
df = pd.read_csv('Covid.csv')
```

```python
df.head()
```

```python
training_data = df.drop('CLASIFFICATION_FINAL', axis=1)
training_label = df['CLASIFFICATION_FINAL']
```

```python
from sklearn.model_selection import cross_validate
from sklearn.metrics import accuracy_score, f1_score, precision_score, recall_score, roc_auc_score, roc_curve, precision_recall_curve, make_scorer, confusion_matrix
scoring_metrics = {
    'accuracy': make_scorer(accuracy_score),
    'precision': make_scorer(precision_score),
    'recall': make_scorer(recall_score),
    'f1': make_scorer(f1_score),
    'roc_auc_score': make_scorer(roc_auc_score),
}
```

```python
from cuml.neighbors import KNeighborsClassifier
folds = 5
for i in range (1, 6):
  knn_model = KNeighborsClassifier(n_neighbors = i, metric="euclidean")
  knn_scores = cross_validate(knn_model, training_data, training_label, cv=folds, scoring=scoring_metrics)

  for i in range(folds):
      print(f"Accuracy for fold {i} = {knn_scores['test_accuracy'][i]}")
      print(f"Precision for fold {i} = {knn_scores['test_precision'][i]}")
      print(f"Recall for fold {i} = {knn_scores['test_recall'][i]}")
      print(f"F1-score for fold {i} = {knn_scores['test_f1'][i]}")
      print(f"AUC for fold {i} = {knn_scores['test_roc_auc_score'][i]}")
      print("\n")

  print("\n", f"Final Metrics for k = {i}>>")
  print(f"Accuracy = {sum(knn_scores['test_accuracy'])/folds}")
  print(f"Precision = {sum(knn_scores['test_precision'])/folds}")
  print(f"Recall = {sum(knn_scores['test_recall'])/folds}")
  print(f"F1-score = {sum(knn_scores['test_f1'])/folds}")
  print(f"AUC = {sum(knn_scores['test_roc_auc_score'])/folds}")
```
